```python
from bs4 import BeautifulSoup
import requests as request
```

```python
url = "https://es.wikipedia.org/wiki/Pirater%C3%ADa"

html = request.get(url).text

bs = BeautifulSoup(html, 'html5lib')

text = bs.get_text(strip=True)

text
```

### Once we have the text, we can tokenize it into sentences and words.

```python
from nltk.tokenize import sent_tokenize, word_tokenize
tokens = word_tokenize(text, language='spanish')
tokens = [word.lower() for word in tokens if word.isalpha()] # remove punctuation, numbers ....

tokens
```

### now we are goint to check the frequency of each word in the text

```python
freq = nltk.FreqDist(tokens)

for key, val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
import seaborn as sns

sns.set()
freq.plot(30, cumulative=False)
```

## now we are going to remove the stop words such as determinations, which are words that are not relevant for the analysis

```python
from nltk.corpus import stopwords
stop_words_spanish = set(stopwords.words('spanish'))
stop_words_english = set(stopwords.words('english'))
stop_words = stop_words_spanish.union(stop_words_english)

clean_tokens = tokens[:]

for token in tokens:
    if token in stop_words:
        clean_tokens.remove(token)

clean_tokens
```

```python
freq = nltk.FreqDist(clean_tokens)

sns.set()
freq.plot(30, cumulative=False)
```

### now we are going to get the synonyms and antonyms of each word using WORDNET

```python
from nltk.corpus import wordnet

synonyms = []
word_to_find_synonyms_en = 'pirate'
word_to_find_synonyms_es = 'pirata'

for syn in wordnet.synsets(word_to_find_synonyms_en):
    for lemma in syn.lemmas():
        if lemma.name() not in synonyms:
            synonyms.append(lemma.name())

print(synonyms)
```

### By default WordNet only returns words in English, so we are going to use the Google Translate API to translate the words to Spanish

```python
synonyms = list(map(lambda synonym: translator.translate(synonym), synonyms))

synonyms
```

```python
clean_tokens_syn = clean_tokens[:]

for index, syn in enumerate(synonyms):
    clean_tokens_syn = [word.replace(synonyms[index], word_to_find_synonyms_es) for word in clean_tokens_syn]
```

```python
manual_synonyms = ['barcos', 'naves']

for index, syn in enumerate(manual_synonyms):
    clean_tokens_syn = [word.replace(manual_synonyms[index], 'barcos') for word in clean_tokens_syn]
```

```python
freq = nltk.FreqDist(clean_tokens_syn)

sns.set()
freq.plot(30, cumulative=False) ## as we can see 'naves' doesn't exist's and 'barcos' is most frequent now
```

### if we want to check the antonyms we can do it in the same way

```python
antonyms = []
word_to_find_antonyms = 'good'

for syn in wordnet.synsets(word_to_find_antonyms):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            print(lemma.antonyms())
            antonyms.append(lemma.antonyms()[0].name())

antonyms
```

```python
antonyms = list(map(lambda antonym: translator.translate(antonym), antonyms))

antonyms
```
