## Why use Derivative Regression?, this method could be used to normalize all gerunds or verbs to his root , for example, running, run, runs, etc. to run.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

print(stemmer.stem("running"))
print(stemmer.stem("runs"))
print(stemmer.stem("run"))
```

#### By default PorterStemmer uses the english language, but we can use other languages, for example spanish with another stemmer.


```python
from nltk.stem import SnowballStemmer

snowball_stemmer = SnowballStemmer("spanish")

print(snowball_stemmer.stem("corriendo"))
print(snowball_stemmer.stem("correr"))
```

```python
import requests as request
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import seaborn as sns

url = "https://es.wikipedia.org/wiki/Pirater%C3%ADa"

html = request.get(url).text

bs = BeautifulSoup(html, 'html5lib')

text = bs.get_text(strip=True)

tokens = word_tokenize(text, language='spanish')
tokens = [word.lower() for word in tokens if word.isalpha()] # remove punctuation, numbers ....

stop_words_spanish = set(stopwords.words('spanish'))
stop_words_english = set(stopwords.words('english'))
stop_words = stop_words_spanish.union(stop_words_english)

clean_tokens = tokens[:]

for token in tokens:
    if token in stop_words:
        clean_tokens.remove(token)
```

```python
clean_tokens_stem = [snowball_stemmer.stem(token) for token in clean_tokens]

freq = nltk.FreqDist(clean_tokens_stem)

sns.set()
freq.plot(20, cumulative=False)
```
