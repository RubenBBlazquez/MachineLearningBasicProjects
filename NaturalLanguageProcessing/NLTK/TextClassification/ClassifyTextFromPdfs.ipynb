```python
import PyPDF2
import nltk
```

```python
pdf_file_object = open('../files/spark_intro.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(pdf_file_object)
```

```python
page_count = len(pdf_reader.pages)
text = ''

for i in range(page_count):
    page = pdf_reader.pages[i]
    text_pag = page.extract_text()

    text += text_pag

text
```

```python
from nltk.tokenize import word_tokenize

tokens_en = word_tokenize(text, 'english')
tokens_es = word_tokenize(text, 'spanish')

words = [*tokens_en, *tokens_es]
words = list(set(words))

tokens = [word.lower() for word in words if word.isalpha()]

tokens
```

```python
from nltk.corpus import stopwords

clean_tokens = tokens[:]

for token in tokens:
    if token in stopwords.words('spanish') or token in stopwords.words('english'):
        clean_tokens.remove(token)
```

```python
import seaborn as sns
import nltk

sns.set()
freq = nltk.FreqDist(clean_tokens)
freq.plot(30, cumulative=False)
```

```python
import spacy
import nltk

nlp_es = spacy.load('es_core_news_sm')
nlp_en = spacy.load('en_core_web_sm')

clean_tokens_lem_es = []

for token in nlp_es(' '.join(clean_tokens)):
    print(token.text, token.lemma_, token.pos_)
    clean_tokens_lem_es.append(token.lemma_)
```
