## Sprectral Clustering

```python
from sklearn.datasets import make_circles

x, y = make_circles(n_samples=1000, noise=0.05, factor=0.5, random_state=42)
```

```python
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=1000, noise=0.05, random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.RdYlBu)
```

```python
from sklearn.cluster import SpectralClustering

clustering = SpectralClustering(n_clusters=2, affinity='rbf', gamma=1)
clustering.fit(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=clustering.labels_, cmap=plt.cm.RdYlBu)
```

```python
clustering = SpectralClustering(n_clusters=2, affinity='rbf', gamma=25)
clustering.fit(x)

plt.scatter(x[:, 0], x[:, 1], c=clustering.labels_, cmap=plt.cm.RdYlBu)
```
